```python
import numpy as np
a = np.arange(0,60,5)
a = a.reshape(3,4)
print("original array is: ")
print(a)
print("\n")
for x in np.nditer(a,op_flags=['readwrite']):
  x[...] = 2*x
print("modified array is ")
print(a)
```

```python
import numpy as np
a = np.arange(0,60,5)
a = a.reshape(3,4)
print("original array is: ")
print(a)
print("\n")
print("second array is :")
b = np.array([1,2,3,4],dtype=int)
print(b)
print("\n")
print("modified array is:")
for x,y in np.nditer([a,b]):
  print("%d:%d"%(x,y))
```

```python
import PIL
import numpy as np
import requests
from io import BytesIO

# URL of a different image
image_url = "https://via.placeholder.com/600x400.png?text=Placeholder+Image"

# Download the image content
response = requests.get(image_url)
response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

# Open the image from the in-memory binary content
img = np.array(PIL.Image.open(BytesIO(response.content)))
print(img.shape)
```

```python
import random,collections,numpy as np

states = ["sunny","rainy"]
trans = [["0.7","0.3"],["0.4","0.6"]]
current = "sunny"
for _ in range(5):
  current = np.random.choice(states,p=trans[states.index(current)])
print(current)
```
