**Case Study 3: Sales Analysis**
--------------------------------------------------------------------------------
Aim

To analyze weekly sales data using NumPy statistical functions.

```python
#Create Dataset
sales = np.array([1200, 1500, 1100, 1700, 1600, 1800, 1400])

print("Weekly Sales Data:", sales)
```

```python
#Average Daily Sales
average_sales = np.mean(sales)
print("Average Daily Sales:", average_sales)
```

```python
#Day with Highest Sales
highest_day = np.argmax(sales)
print("Highest Sales on Day:", highest_day + 1)
```

```python
#Percentage Growth (Day 1 to Day 7)
percentage_growth = ((sales[-1] - sales[0]) / sales[0]) * 100
print("Percentage Growth from Day 1 to Day 7:", percentage_growth, "%")
```

```python
#Cumulative Sales
cumulative_sales = np.cumsum(sales)
print("Cumulative Sales:", cumulative_sales)
```

```python
#Standard Deviation
sales_std = np.std(sales)
print("Standard Deviation of Sales:", sales_std)
```

**Interpretation of case 3**
--------------------------------------------------------------------------------

Average daily sales indicate steady revenue generation.

Sales peaked on Day 6.

There is overall growth from Day 1 to Day 7.

Standard deviation suggests moderate variation in daily sales performance.
