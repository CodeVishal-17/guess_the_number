## Task 2: Indexing & Slicing (1 Mark)

Given:

A = np.arange(1, 26).reshape(5,5)

Perform:

Extract 2nd row

Extract 3rd column

Extract 2×2 submatrix (center)

Reverse last row

```python
import numpy as np

A = np.arange(1, 26).reshape(5,5)
print(A)
print("\n")

# Extract 2nd row
print(A[1])
print("\n")

# Extract 3rd column
print(A[:, 2])
print("\n")

# Extract 2x2 center matrix
print(A[1:3, 1:3])
print("\n")

# Reverse last row
print(A[-1][::-1])
print("\n")
```

## Task 3: Broadcasting(1 Mark)

Perform broadcasting operations on matrix A created in Task-2.

Add scalar 10 to matrix A

Add vector [1,2,3,4,5] to each row

Normalize matrix values between 0–1

```python
import numpy as np

A = np.arange(1, 26).reshape(5,5)
print(A)

# Perform broadcasting operations here

# Add scalar 10 to matrix
print("\nAdd scalar 10 to matrix:\n", A + 10)

# Add vector [1,2,3,4,5] to each row
print("\nAdd vector [1,2,3,4,5] to each row:", A + np.array([1,2,3,4,5]))

# Normalize matrix between 0 and 1
normalized_A = (A - A.min()) / (A.max() - A.min())
print("\nNormalized matrix between 0 and 1:", normalized_A)
```

## Task 4: Vectorized Computation(1 Marks)

Perform element-wise multiplication
Compute dot product
Compute mean, standard deviation

```python
import numpy as np

X = np.array([1,2,3,4])
Y = np.array([5,6,7,8])

# Element-wise multiplication
print(f"Element-wise multiplication:{X*Y}\n")

# Dot product
x_y_dot = np.dot(X,Y)
print(f"Dot product:{x_y_dot}\n")

# Mean and Standard Deviation
x_mean = np.mean(X)
print(f"Mean of X:{x_mean}\n")
x_standard = np.std(X)
print(f"Standard Deviation of X:{x_standard}\n")
y_mean = np.mean(Y)
print(f"Mean of Y:{y_mean}\n")
y_standard = np.std(Y)
print(f"Standard Deviation of Y:{y_standard}\n")
```

Task 5: Mini Case Problem (1 Mark)

Dataset: Student marks (5 students, 3 subjects)

Compute total and average per student

Find topper

Find subject-wise highest score

```python
import numpy as np

marks = np.array([[80,75,90],[60,88,70],[95,92,85],[70,65,60],[88,78,84]])

# Compute totals, averages, topper

# Total marks per student
totals = np.sum(marks, axis=1)
print("Total marks per student:", totals)

# Average marks per student
averages = np.mean(marks, axis=1)
print("Average marks per student:", averages)

# Find topper (student with highest total)
topper_index = np.argmax(totals)
print("Topper is student index:", topper_index)
print("Topper total marks:", totals[topper_index])

# Subject-wise highest score
subject_highest = np.max(marks, axis=0)
print("Subject-wise highest scores:", subject_highest)
```
